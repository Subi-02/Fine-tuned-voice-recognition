```python
import zipfile
import os

zip_path = "RAVDESS/Audio_Speech_Actors_01-24.zip"
extract_path = "RAVDESS/Audio_Speech_Actors_01-24"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Check contents
os.listdir(extract_path)
```

```python
base_path = "RAVDESS/Audio_Speech_Actors_01-24"

for actor in os.listdir(base_path):
    actor_path = os.path.join(base_path, actor)
    for file in os.listdir(actor_path):
        print(file)  # check audio files
```

```python
import glob

audio_paths = glob.glob("RAVDESS/**/*.wav", recursive=True)
labels = []

# RAVDESS emotion mapping
emotion_map = {
    '01': 'neutral',
    '02': 'calm',
    '03': 'happy',
    '04': 'sad',
    '05': 'angry',
    '06': 'fearful',
    '07': 'disgust',
    '08': 'surprised'
}

for path in audio_paths:
    filename = os.path.basename(path)
    emotion_code = filename.split('-')[2]
    labels.append(emotion_map[emotion_code])

print("Total audio files:", len(audio_paths))
print("Sample:", audio_paths[:3], labels[:3])
```

```python
from transformers import Wav2Vec2Processor

processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
from datasets import Dataset

dataset = Dataset.from_dict({
    "path": audio_paths,
    "emotion": labels
})

print(dataset)
```

```python
import torch
import librosa

def preprocess(batch):
    audio, sr = librosa.load(batch["path"], sr=16000)
    batch["input_values"] = processor(audio, sampling_rate=16000).input_values[0]
    batch["labels"] = batch["emotion"]
    return batch

dataset = dataset.map(preprocess)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
dataset = dataset.map(lambda batch: {"labels": le.fit_transform([batch["labels"]])[0]})
num_labels = len(le.classes_)

print("Number of emotion classes:", num_labels)
```

```python
dataset = dataset.train_test_split(test_size=0.2)
train_ds = dataset["train"]
test_ds = dataset["test"]
```

```python
from transformers import Wav2Vec2ForSequenceClassification

model = Wav2Vec2ForSequenceClassification.from_pretrained(
    "facebook/wav2vec2-base-960h",
    num_labels=num_labels,
    problem_type="single_label_classification"
)
```

```python
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(processor)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./wav2vec2-emotion",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    learning_rate=2e-5,
    logging_steps=10,
    save_steps=500,
    save_total_limit=2,
    fp16=True,
    report_to="none"
)
```

```python
from transformers import (
    Wav2Vec2Processor,
    Wav2Vec2ForSequenceClassification
)
```

```python
emotion_labels = ["neutral", "happy", "sad", "angry"]
num_labels = len(emotion_labels)
```

```python
processor = Wav2Vec2Processor.from_pretrained(
    "facebook/wav2vec2-base"
)
```

```python
model = Wav2Vec2ForSequenceClassification.from_pretrained(
    "facebook/wav2vec2-base",
    num_labels=num_labels,
    problem_type="single_label_classification"
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    tokenizer=processor,
    data_collator=data_collator
)
```

```python
trainer.train()
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
def predict_emotion(audio_path):
    audio, sr = librosa.load(audio_path, sr=16000)

    inputs = processor(
        audio,
        sampling_rate=16000,
        return_tensors="pt",
        padding=True
    )

    # 🔑 MOVE INPUTS TO SAME DEVICE AS MODEL
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)

    predicted_label = torch.argmax(outputs.logits, dim=-1).item()

    emotion_labels = ["neutral", "happy", "sad", "angry"]
    return emotion_labels[predicted_label]
```

```python
model.to(device)
model.eval()
```

```python
test_audio = audio_paths[5]
print("Predicted Emotion:", predict_emotion(test_audio))
```
